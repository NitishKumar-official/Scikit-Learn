# Simple Imputer

```python
df
```

```python
from sklearn.impute import SimpleImputer
```

# mean

```python
simple_mean = SimpleImputer(strategy='mean')
df['A'] = simple_mean.fit_transform(df[['A']])
```

```python
df
```

# median

```python
simple_mean = SimpleImputer(strategy='median')
df['B'] = simple_mean.fit_transform(df[['B']])
```

```python
df
```

# most_frequent

```python
simple_mean = SimpleImputer(strategy='most_frequent')
df['C'] = simple_mean.fit_transform(df[['C']])
```

```python
df
```
